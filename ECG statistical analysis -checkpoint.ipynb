```python
from scipy.stats import pearsonr
```

```python
df=pd.read_excel(r'C:\Users\L user\.ipython\50 Donkey ECG sheet.xlsx')
```

```python
age = df['age']
HR = df['HR']
```

```python
r_age_hr, p_age_hr = pearsonr(age, HR)
round(r_age_hr, 3), round(p_age_hr, 3)
print(f"r = {r_age_hr:.3f}, p = {p_age_hr:.3f}")
```

```python
ECGhr = df['ECGhr']
```

```python
r_age_ecg, p_age_ecg = pearsonr(age, ECGhr)
round(r_age_ecg, 3), round(p_age_ecg, 3)
print(f"r = {r_age_ecg:.3f}, p = {p_age_ecg:.3f}")
```

```python
There was no statistically significant correlation between age and heart rate (r = -0.04, p = 0.77), 
nor between age and ECG-derived heart rate (r = 0.11, p = 0.46).
```

```python
ecg_columns = [
    'Pd', 'QRS d', 'Td', 'PR int', 'QT int',
    'RR int', 'PR seg', 'ST seg',
    'P am', 'QRS am', 'T am'
]
results = []

for col in ecg_columns:
    r, p = pearsonr(df['age'], df[col])
    results.append([col, r, p])
    results_df = pd.DataFrame(
    results,
    columns=['age vs other columns', 'Correlation (r)', 'P-value']
)

results_df

results_df['Correlation (r)'] = results_df['Correlation (r)'].round(3)
results_df['P-value'] = results_df['P-value'].round(3)

results_df
```

```python
Age showed no statistically significant correlation with ECG durations, intervals, or amplitudes (p > 0.05 for all parameters).
```

```python
from scipy.stats import f_oneway
```

```python
groups = []
for work, data in df.groupby('work type'):
    groups.append(data['ECGhr'])
f_stat, p_value= f_oneway(*groups)
f_stat.round(3), p_value.round(3)
print(f"One-way ANOVA result: F = {f_stat:.3f}, p = {p_value:.3f}")
```

```python
One-way ANOVA showed no significant difference in ECG heart rate among different work types (F = 0.124, p = 0.883).
```

```python
ecg_columns = df.loc[:, 'ECGhr':].select_dtypes(include='number').columns
from scipy.stats import f_oneway

results = []

for col in ecg_columns:
    groups = []
    for work, data in df.groupby('work type'):
        groups.append(data[col])
    
    f_stat, p_value = f_oneway(*groups)
    results.append([col, round(f_stat, 3), round(p_value, 3)])
    anova_df = pd.DataFrame(
    results,
    columns=['work type vs other columns', 'F value', 'P-value']
)

anova_df
```

```python
One-way ANOVA revealed no significant differences in most ECG parameters across different work types. However,
T wave amplitude showed a significant difference (F = 5.958, p = 0.005), while QRS amplitude demonstrated a near-significant trend (p = 0.054).
```

```python
df['feed'] = df['feed'].str.strip().str.lower()
df['feed'].value_counts()
```

```python
groups = []

for feed_type, data in df.groupby('feed'):
    groups.append(data['ECGhr'])
f_stat, p_value = f_oneway(*groups)
f_stat.round( 3), p_value.round(3)
print(f"One-way ANOVA result: F = {f_stat:.3f}, p = {p_value:.3f}")
```

```python
ecg_columns = df.columns[df.columns.get_loc('ECGhr') + 1:]
ecg_columns = df[ecg_columns].select_dtypes(include=['float64', 'int64']).columns
```

```python
results = []
for col in ecg_columns:
    groups = []
    
    for feed_type, data in df.groupby('feed'):
        groups.append(data[col])
    
    f_stat, p_value = f_oneway(*groups)
    results.append([col, round(f_stat, 3), round(p_value, 3)])

results_df = pd.DataFrame(
    results,
    columns=['feed vs other columns', 'F value', 'P-value']
)

results_df
```

```python
PR interval showed a trend toward statistical significance among different feed types
however, the difference did not reach the conventional level of significance (p = 0.060).
A statistically significant difference was detected in RR interval among different feed types (F = 3.887, p = 0.015).
```

```python
df['remarks'] = df['remarks'].fillna('no_remarks')
df[['age type', 'remarks']].dtypes
```

```python
contingency_table = pd.crosstab(df['age type'], df['remarks'])
contingency_table
```

```python
from scipy.stats import chi2_contingency
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square = {chi2:.3f}")
print(f"P-value = {p_value:.3f}")
if p_value < 0.05:
    print("There is a statistically significant correlation between age type and remarks")
else:
    print("There is no statistically significant association between age type and remarks")
```
